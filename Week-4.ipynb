## Data Visualization Tasks: Titanic Dataset

### Day 1: Introduction to Matplotlib

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Titanic Dataset
titanic = sns.load_dataset('titanic')

# Line Plot: Age of passengers vs. Index
plt.figure(figsize=(8, 4))
plt.plot(titanic.index, titanic['age'], color='blue')
plt.title('Passenger Ages Over Indices')
plt.xlabel('Passenger Index')
plt.ylabel('Age')
plt.show()

# Scatter Plot: Age vs. Fare
plt.figure(figsize=(8, 4))
plt.scatter(titanic['age'], titanic['fare'], alpha=0.5, color='green')
plt.title('Scatter Plot: Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

# Bar Chart: Survival count
survival_counts = titanic['survived'].value_counts()
plt.figure(figsize=(6, 4))
plt.bar(survival_counts.index, survival_counts.values, color=['red', 'green'])
plt.title('Survival Count')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.ylabel('Count')
plt.show()
```

### Day 2: Customizing Plots in Matplotlib

```python

plt.figure(figsize=(8, 4))
plt.bar(['Male', 'Female'], titanic['sex'].value_counts(), color=['blue', 'pink'])
plt.title('Passenger Count by Gender', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.annotate('Highest', xy=(0, 600), xytext=(0, 700), 
             arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10)
plt.show()
```

### Day 3: Exploring Seaborn Basics

```python

# Distplot: Fare Distribution
sns.histplot(titanic['fare'], kde=True, color='orange', bins=30)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()

# KDE Plot: Age
sns.kdeplot(titanic['age'], shade=True, color='purple')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()

# Count Plot: Embarkation Ports
sns.countplot(x='embark_town', data=titanic, palette='viridis')
plt.title('Passenger Count by Embarkation Port')
plt.show()
```

### Day 4: Exploring Categorical Data

```python

# Boxplot: Fare by Class
sns.boxplot(x='class', y='fare', data=titanic, palette='pastel')
plt.title('Boxplot of Fare by Class')
plt.show()

# Violinplot: Age by Gender
sns.violinplot(x='sex', y='age', data=titanic, split=True, palette='muted')
plt.title('Violin Plot of Age by Gender')
plt.show()

# Stripplot: Age by Survival
sns.stripplot(x='survived', y='age', data=titanic, jitter=True, hue='sex', palette='Set2', dodge=True)
plt.title('Age Distribution by Survival Status')
plt.legend(title='Gender')
plt.show()
```

### Day 5: Pairplot and Heatmap

```python

# Pairplot: Numerical Columns
sns.pairplot(titanic[['age', 'fare', 'survived']], hue='survived', palette='coolwarm')
plt.suptitle('Pairplot of Age, Fare, and Survival', y=1.02)
plt.show()

# Heatmap: Correlation Matrix
correlation_matrix = titanic[['age', 'fare', 'survived']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

### Day 6: Bar and Count Charts

```python

# Grouped Bar Chart: Survival by Class
survival_by_class = titanic.groupby(['class', 'survived']).size().unstack()
survival_by_class.plot(kind='bar', stacked=True, figsize=(8, 6), color=['red', 'green'])
plt.title('Survival by Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.legend(['Did Not Survive', 'Survived'])
plt.show()

# Count Chart: Passengers by Age Group
titanic['age_group'] = pd.cut(titanic['age'], bins=[0, 18, 40, 60, 80], labels=['Child', 'Young Adult', 'Adult', 'Senior'])
sns.countplot(x='age_group', data=titanic, hue='survived', palette='viridis')
plt.title('Passenger Count by Age Group and Survival')
plt.show()
```

### Day 7: Subplots and Advanced Layouts

```python

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: Fare Distribution
sns.histplot(titanic['fare'], kde=True, color='blue', ax=axes[0, 0])
axes[0, 0].set_title('Fare Distribution')

# Plot 2: Age Distribution
sns.kdeplot(titanic['age'], shade=True, color='orange', ax=axes[0, 1])
axes[0, 1].set_title('Age Distribution')

# Plot 3: Survival Count
sns.countplot(x='survived', data=titanic, palette='coolwarm', ax=axes[1, 0])
axes[1, 0].set_title('Survival Count')

# Plot 4: Embarkation Town Count
sns.countplot(x='embark_town', data=titanic, palette='Set2', ax=axes[1, 1])
axes[1, 1].set_title('Passengers by Embarkation Town')

plt.tight_layout()
plt.show()
```
