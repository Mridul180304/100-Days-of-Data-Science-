# Week 6: Unsupervised Learning

This notebook focuses on unsupervised learning techniques, where the objective is to identify patterns, structures, or groupings in data without any labeled outcomes. Unlike supervised learning, unsupervised learning does not use predefined labels, making it ideal for clustering, dimensionality reduction, and anomaly detection.

### Techniques Covered:
- **K-Means Clustering:** Partitions data into a predefined number of clusters.
- **Hierarchical Clustering:** Builds a tree of clusters based on similarity.
- **DBSCAN:** Groups data points based on density, identifying noise and clusters of arbitrary shapes.
- **Principal Component Analysis (PCA):** Reduces the dimensionality of data while preserving variance.
### Weekly Breakdown:
- **Day 1-2:** K-Means Clustering
- **Day 3:** Hierarchical Clustering
- **Day 4-5:** DBSCAN Clustering
- **Day 6-7:** PCA (Principal Component Analysis)

Let's dive into these techniques with code examples and visualizations!

```python

# Day 1-2: K-Means Clustering
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot generated data
axes[0].scatter(X[:, 0], X[:, 1], s=50)
axes[0].set_title("Generated Data")
axes[0].set_xlabel("Feature 1")
axes[0].set_ylabel("Feature 2")

# Apply K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Plot K-Means clustering result
axes[1].scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50)
centers = kmeans.cluster_centers_
axes[1].scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X', label='Centroids')
axes[1].set_title("K-Means Clustering")
axes[1].set_xlabel("Feature 1")
axes[1].set_ylabel("Feature 2")
axes[1].legend()

plt.subplots_adjust(wspace=0.3)
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# Generate synthetic data (reuse X)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))  # Set figsize for the entire figure

# Plot generated data
axes[0].scatter(X[:, 0], X[:, 1], s=50)
axes[0].set_title("Generated Data")
axes[0].set_xlabel("Feature 1")
axes[0].set_ylabel("Feature 2")

# Create Dendrogram
linked = linkage(X, method='ward')
dendrogram(linked, ax=axes[1])
axes[1].set_title("Dendrogram")
axes[1].set_xlabel("Feature 1")
axes[1].set_ylabel("Feature 2")

# Apply Hierarchical Clustering
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)

# Plot the clusters
axes[2].scatter(X[:, 0], X[:, 1], c=y_hc, cmap='rainbow', s=50)
axes[2].set_title("Hierarchical Clustering")
axes[2].set_xlabel("Feature 1")
axes[2].set_ylabel("Feature 2")

plt.subplots_adjust(wspace=0.3)
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# Generate synthetic data (reuse X)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot generated data
axes[0].scatter(X[:, 0], X[:, 1], s=50)
axes[0].set_title("Generated Data for DBSCAN")
axes[0].set_xlabel("Feature 1")
axes[0].set_ylabel("Feature 2")

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

# Plot DBSCAN clustering result
axes[1].scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='plasma', s=50)
axes[1].set_title("DBSCAN Clustering")
axes[1].set_xlabel("Feature 1")
axes[1].set_ylabel("Feature 2")

plt.subplots_adjust(wspace=0.3)
plt.show()
```

```python

# Day 6-7: Principal Component Analysis (PCA)
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Generate high-dimensional synthetic data
X_hd, _ = make_blobs(n_samples=300, centers=4, cluster_std=5, n_features=5, random_state=42)

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_hd)

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot the transformed data
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', s=50)
plt.title("PCA: Reduced to 2D")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

plt.show()
```
