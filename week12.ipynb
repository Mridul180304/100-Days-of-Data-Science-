## Day 1: Descriptive Statistics


Learn the basics of descriptive statistics, such as measures of central tendency and variability.

### Concepts:
- Mean, Median, Mode
- Standard Deviation, Variance
- Minimum, Maximum, Range


```python

import numpy as np
import pandas as pd

# Sample data
data = [12, 15, 14, 10, 8, 13, 14, 14, 12, 10]

# Using NumPy
mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)
variance = np.var(data)
minimum = np.min(data)
maximum = np.max(data)
data_range = maximum - minimum

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_dev}")
print(f"Variance: {variance}")
print(f"Range: {data_range}")

# Using Pandas
df = pd.DataFrame(data, columns=['Values'])
print("\nPandas Descriptive Statistics:")
print(df.describe())
```

## Day 2: Inferential Statistics â€“ Sampling


Understand random sampling and how it influences statistical inference.

### Concepts:
- Population vs. Sample
- Simple Random Sampling
- Sampling Distribution


```python

from scipy.stats import norm
import matplotlib.pyplot as plt

# Population data
population = np.random.normal(loc=50, scale=10, size=10000)

# Random sample
sample = np.random.choice(population, size=100, replace=False)

# Visualize population vs. sample
plt.hist(population, bins=50, alpha=0.5, label='Population')
plt.hist(sample, bins=20, alpha=0.8, label='Sample')
plt.legend()
plt.title("Population vs. Sample")
plt.show()

# Sample mean and standard deviation
print(f"Sample Mean: {np.mean(sample)}")
print(f"Sample Standard Deviation: {np.std(sample)}")
```

## Day 3: Hypothesis Testing


Learn to test hypotheses using statistical tests.

### Concepts:
- Null and Alternative Hypothesis
- t-Test
- p-Value


```python

from scipy.stats import ttest_1samp

# One-sample t-test
sample_data = np.random.normal(50, 10, size=30)  # Sample data
population_mean = 50

t_stat, p_value = ttest_1samp(sample_data, population_mean)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

## Day 4: Correlation and Covariance


Explore relationships between variables.

### Concepts:
- Covariance
- Pearson Correlation
- Spearman Rank Correlation


```python

from scipy.stats import pearsonr, spearmanr

# Data
x = np.random.rand(100)
y = 2 * x + np.random.normal(0, 0.1, size=100)

# Covariance matrix
cov_matrix = np.cov(x, y)
print(f"Covariance matrix:\n{cov_matrix}")

# Pearson correlation
pearson_corr, _ = pearsonr(x, y)
print(f"Pearson Correlation: {pearson_corr}")

# Spearman correlation
spearman_corr, _ = spearmanr(x, y)
print(f"Spearman Rank Correlation: {spearman_corr}")
```

## Day 5: Probability Distributions


Learn about probability distributions and their importance in statistics.

### Concepts:
- Normal Distribution
- Binomial Distribution
- Poisson Distribution


```python

from scipy.stats import norm, binom, poisson
import matplotlib.pyplot as plt

# Normal distribution
x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x, loc=0, scale=1), label='Normal Distribution')

# Binomial distribution
n, p = 10, 0.5
x_binom = np.arange(0, n+1)
plt.bar(x_binom, binom.pmf(x_binom, n, p), alpha=0.6, label='Binomial Distribution')

# Poisson distribution
x_poisson = np.arange(0, 20)
plt.bar(x_poisson, poisson.pmf(x_poisson, mu=5), alpha=0.6, label='Poisson Distribution')

plt.legend()
plt.title("Probability Distributions")
plt.show()
```

## Day 6: Regression Analysis


Explore linear regression and its statistical interpretation.

### Concepts:
- Simple Linear Regression
- Coefficient of Determination (RÂ²)


```python

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Generate data
np.random.seed(42)
X = np.random.rand(100, 1) * 10
y = 5 * X + np.random.normal(0, 3, (100, 1))

# Fit linear regression model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Model parameters
print(f"Intercept: {model.intercept_[0]}")
print(f"Coefficient: {model.coef_[0][0]}")
print(f"RÂ² Score: {r2_score(y, y_pred)}")

# Plot
plt.scatter(X, y, label='Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.legend()
plt.title("Linear Regression")
plt.show()
```
